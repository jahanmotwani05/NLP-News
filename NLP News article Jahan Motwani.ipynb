# Web Scraping for News Article


```python
#Importing necessary libraries for web scraping 
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
from urllib.request import urlopen

#Information of link of website containing news article 

#my_url="https://blog.google/products/search/google-search-new-fact-check-features-europe/"

#You can ask the user to give input also
my_url=input("Enter the link for doing web scraping")

#Open the url with urlopen() function
urlinfo=urlopen(my_url)

#Reading data from news article in html format
urlhtml = urlinfo.read()

urlinfo.close()
```

```python
#displaying the html file of the link
print(urlhtml)
```

```python
#Extracting the content from the html which has div tag and the class is "uni-paragraph article-paragraph"
page_soup=soup(urlhtml,"html.parser")

#classname="uni-paragraph article-paragraph"

classname=input("Please enter the name of the class from where the web scraping needs to be done: ")

articles=page_soup.findAll("div",{"class":classname})

#Displaying the number of div tags having class as "uni-paragraph article-paragraph"
print(len(articles))
```

```python
#Displaying the first and second div tag
for i in range(len(articles)):
    print(articles[i])
```

```python
news_articles=""
for article in articles:
    print(article.text)
    news_articles=news_articles.join(article.text)
```

```python
#Just taking the first article of news
#Storing the content using text and displaying the text data
textdata=articles[0].text
print(textdata)
```

# Text Pre-Processing for removing special characters, punctuation marks


```python
#Importing necessary libraries
from string import punctuation
from collections import Counter

#Use of translate() function to remove punctuation marks in document. 
#Use of lower() function to convert text in lowercase.
#Use of split() function to spilt the document in words.
word_doc = textdata.translate(punctuation).lower().split()
print("--------Converting document into lower case words-----")
print(word_doc)
print("The number of words in the document are: ",len(word_doc))

#Use of Counter() function to display the most common words.
word_counts = Counter(word_doc)
print("The most Common words in document are: ")
for word in word_counts.most_common(11):
    print(word)
```

```python
#Program related to Data Cleaning 
#Importing Necessary Libraries
import nltk
import re

#Determining number of letters in the original document.
print("Number of letters in the original document:", len(textdata))

#Removing unwanted characters from document using regular expressions.
textdoc=re.sub(r'[^a-zA-Z0-9\s]','',textdata,re.I|re.A)

#Determining number of letters in the new document.
print("Number of letters after removing unwanted characters:", len(textdoc))

#Converting all the words to lower case for exact determination.
textdoc=textdoc.lower()

#Removing leading and trailing spaces from document using strip().
textdoc=textdoc.strip()

#Displaying the Document
print("The new document is:\n",textdoc)
```

# Creating sentence and word tokens from text data 

```python
#Sentence Tokenization using sent_tokenize() function
from nltk.tokenize import sent_tokenize, word_tokenize
sent_tokens=nltk.sent_tokenize(textdata)
print("The sentence tokens are:\n",sent_tokens)

#Tokenize document into tokens/words using word_tokenize() function.
doc_tokens=nltk.word_tokenize(textdoc)
print("The word tokens are:\n",doc_tokens)
```

# Removing stopwords from the document

```python
#Downloading stop words from nltk library
nltk.download('stopwords')

#Storing all stop words of English from nltk corpus. 
eng_stopwords=nltk.corpus.stopwords.words('english')

#Determining the number of English stop words in corpus.
print("Number of English stop words in corpus:", len(eng_stopwords))

#Displaying the first 30 stop words. 
print("Example of first 30 stop words include:\n",eng_stopwords[0:30])

#Filtering the tokens (not in list of stop words) from list of tokens 
final_tokens=[token for token in doc_tokens
                        if token not in eng_stopwords]

#Joining the filtered tokens using join() function.
newdoc=' '.join(final_tokens)

#Determining number of letters after removal of stop words in document.
print("Number of letters after removal of stop words:",len(newdoc))
print("First 298 letters from my new document are: \n",newdoc[0:298])

```

# Performing lemmatization and Stemming

```python
#Performing stemming and lemmatization on data
#Importing necessary libraries
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
ps = PorterStemmer() 
lem=WordNetLemmatizer() 
lemlist=[]
pslist=[]

#Creating a list of words having same base word
for word in final_tokens:
    lemlist.append(lem.lemmatize(word,'v'))
    pslist.append(ps.stem(word))

#Displaying the result after Lemmatization and Stemming
print("After Lemmatization:\n",lemlist[0:20])
print("After Stemming:\n",pslist[0:20])
```

# Creating wordcloud for displaying the summary of the document

```python
#Importing necessary libraries for creating visual display of data
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 

#Creating a word cloud of words without lemmatization
start_all_words=''
stopwords = set(STOPWORDS) 
for word in doc_tokens: 
    start_all_words = start_all_words + word + ' '
 
start_wordcloud = WordCloud(width = 1000, height = 1000, 
                background_color ='white', stopwords = stopwords, 
                min_font_size = 10).generate(start_all_words) 

#Creating a word cloud of words with lemmatization
clean_all_words=''
for newword in lemlist: 
    clean_all_words = clean_all_words + newword + ' '
 
clean_wordcloud = WordCloud(width = 1000, height = 1000, 
                background_color ='black', stopwords = stopwords, 
                min_font_size = 10).generate(clean_all_words)  

#Plot both the wordcloud images                        
plt.figure(1, figsize=(20,10))
plt.subplot(121)
plt.imshow(start_wordcloud) 
plt.subplot(122)
plt.imshow(clean_wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show()
```
